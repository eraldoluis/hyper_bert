```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set(style="whitegrid")
plt.clf()

df = pd.read_csv("../results/subtoken_size/subtoken_dataset.tsv", delimiter="\t")
dataset_names = df.dataset.unique().tolist()
df2 = pd.DataFrame({"subtoken": ["1,1", "2,1", "3,1", "2,2","3,3", "4,2"], "N": [2,4,7,1,10,5], "dataset": dataset_names * 2})
for d_name in dataset_names:
    plt.figure(figsize=(25,8))
    ax = sns.histplot(x="subtoken", weights="N", stat="count", discrete=True, data=df[df['dataset'] == d_name])
    sum = df[df['dataset'] == d_name]['N'].sum()
    ax.set(xlabel="Tamanhos dos hipônimos e hiperônimos")
    ax.set_title(f"")
    ax.set_title(f"Dataset {d_name} com {sum} pares")
    plt.show()

```
